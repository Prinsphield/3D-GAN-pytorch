```python
import os
import scipy.ndimage as nd
import scipy.io as io
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

%matplotlib notebook


# import skimage.measure as sk

def plotFromVoxels(voxels):
    z,x,y = voxels.nonzero()
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x, y, -z, zdir='z', c= 'red')
    plt.show()

def getVoxelFromMat(path, cube_len=64):
    voxels = io.loadmat(path)['instance']
    voxels = np.pad(voxels,(1,1),'constant',constant_values=(0,0))
    if cube_len != 32 and cube_len == 64:
        voxels = nd.zoom(voxels, (2,2,2), mode='constant', order=0)
    return voxels

filename = '/gpfs/share/home/1501210096/datasets/ModelNet/volumetric_data/airplane/30/test/fecbd7475115dc85237c15a579990af3_9.mat'
voxels = getVoxelFromMat(filename)

# plotFromVoxels(voxels)
print(voxels.shape)

# filename = 'train_log/img/002000_04.mat'
# f = io.loadmat(filename)
# print(f)
# # print(f['instance'].shape)


save_file = 'save.mat'
io.savemat(save_file, {'instance': voxels})
```
